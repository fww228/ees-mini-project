```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### Read in the data

```python
dataset = pd.read_csv('temp.csv')
print('Dataset features:', dataset.columns, '\n')
print('Head 5 of the dataset:')
print(dataset.head())
print()
print('Dataset description:')
print(dataset.info())
```

From the dataset description we can see that there are 43464 samples and no missing values.

Meanwhile, this dataset has 8 features, where:

- `Year`, `Month`, `Day` and `Date` describe the specific date of the data.
- `PPT.` describes the daily rainfall.
- `Av temp` describes the daily average temperature, which is also the target.
- `Tmax` describes the daily maximum temperature.
- `Tmin` describes the daily minimum temperature.

### Clean the data

According to the question, we need to predict the daily average temperature in 2020, so feature `Date`, `Tmax` and `Tmin` from previous years are not important to us, then they are discarded.

```python
dataset = dataset.drop(columns='Date')
dataset = dataset.drop(columns='Tmax')
dataset = dataset.drop(columns='Tmin')
```

Finding the relationship between the daily rainfall and the daily average temperature through the data from the previous 10 years.

```python
sns.pairplot(
    dataset[:3650],
    x_vars=['PPT. ', 'Av temp'],
    y_vars=['PPT. ', 'Av temp'],
    )
```

The graphs show that there is no relationship between the daily rainfall and the daily average temperature, which can be removed directly.

```python
dataset = dataset.drop(columns='PPT. ')
dataset.head()
```

### Using Prophet to predict the data

Loading the prophet package and setting the parameters.

```python
from prophet import Prophet
dates = [str(year) + '-' + str(month) + '-' + str(day) for year, month, day in zip(dataset['Year'], dataset['Month'], dataset['Day'])]
dates = pd.to_datetime(dates)
temps = dataset['Av temp']
```

Fitting the data to the prophet model.

```python
df = pd.DataFrame({'ds': dates, 'y': temps})
pp_model = Prophet().fit(df)
```

Using the model to predict the temporature of the future 366 days (year 2020).

```python
future_temps = pp_model.make_future_dataframe(periods=366)
forecast = pp_model.predict(future_temps)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

Plotting the whole data and the predicted data.

```python
plt.figure(figsize=(16, 4))
plt.plot(forecast['ds'], forecast['yhat'], label='Prediction')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.show()
```

Plotting the predicted data only.

```python
plt.figure(figsize=(10, 4))
plt.plot(forecast['ds'][-366:], forecast['yhat'][-366:], label='Prediction')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.show()
```

Plotting the trend of the predicted data.

```python
fig2 = pp_model.plot_components(forecast)
```

### Diagnostics

Using cross validation to calculate the error.

```python
from prophet.diagnostics import cross_validation
from prophet.diagnostics import performance_metrics
df_cv = cross_validation(pp_model, initial='730 days', period='365 days', horizon='365 days')
df_metrics = performance_metrics(df_cv)
```

Have a look at the errors.

```python
df_metrics.head()
```

Plotting the RMSE, from which we can see that the error is low.

```python
from prophet.plot import plot_cross_validation_metric
fig_rmse = plot_cross_validation_metric(df_cv, metric='rmse', rolling_window=0.1)
```
